```python
df.describe().T
```

```python
#quando coloca para incluir todas, ele puxa todos mas no consegue fazer a estatistica deles por no ser inteiro.
df.describe(include='all').T
```

```python
# convertendo o tipo de dados para data
df['Last Payment Date'] = pd.to_datetime(df['Last Payment Date'])
df['Join Date'] = pd.to_datetime(df['Join Date'])
```

```python
df.info()
```

```python
# h quanto tempo as pessoas esto inscritas na Netflix?
df['duracao_assinatura'] = df['Last Payment Date'] - df['Join Date']
df['duracao_assinatura'] = df['duracao_assinatura'].dt.days
df
```

```python
# Com base na funo describe() percebemos que a mdia de tempo que as pessoas se inscrevem  de cerca de 326 dias, quase um ano.
# Mas, pera a, o que so esse nmeros negativos no minimo?
df.describe().T
```

```python
# vamos validar esses numeros?
df[(df['duracao_assinatura'] < 0)]
```

```python
numero_usuarios_negativos = df[df['duracao_assinatura'] < 0]['User ID'].count()
print(f"Nmero de usurios com durao negativa: {numero_usuarios_negativos}")
```

```python
13/2500
```

```python
# Encontramos valores inconsistentes, onde o dado no faz sentido. Como no possumos contexto suficiente sobre essa tabela para entender
# o motivo de termos dt_ultimo_pagamento maior que dt_assinatura, iremos retirar essas linhas para no enviezar nossos resultados.
df = df.drop(df[df['duracao_assinatura'] < 0].index)
df
```

```python
# Agora sim, tirando os dados negativos, nossa mdia de durao de assinatura subiu um pouco, e agora podemos visualizar que o tempo mnimo
# que as pessoas assinam a netflix  de 8 dias, e o mximo da base  776 dias.
df.describe().T
```

```python
df.info()
```

```python
# Antes de continuarmos nossa anlise, existe mais um passo que devemos fazer: resetar o index. Como retiramos algumas linhas da nossa tabela, o
# ndice, ou seja, coluna mais  esquerda, porem ela ainda est com o intervalo de 1-2499.
df = df.reset_index(drop=True)
df

# agora sim! nossa coluna de index est coerente. Vamos continuar!
```

```python
# outra anlise interessante de se fazer, ser que existe um ms com um maior nmero de vendas?
df['dt_mes'] = df['Join Date'].dt.month
df['dt_mes'].value_counts()
```

```python
# Utilizando a funo value_counts() conseguimos ter uma ideia de qual  a assinatura mais comum, porm, em nmeros absolutos
#  dificil entender a proporo entre os 3 planos.
df["Subscription Type"].value_counts()
```

```python
# Para isso, colocamos um parmetro dentro da nossa funo, e ele nos devolve os valores em porcentagem sobre o total.
df["Subscription Type"].value_counts(normalize=True)

# daqui, percebemos que o pacote mais recorrente  o Basic.
```

```python
# Vamos entender tambm, quais sao os pases com o maior nmero de assinaturas: em valor absoluto
df["Country"].value_counts()
```

```python
# Vamos entender tambm, quais sao os pases com o maior nmero de assinaturas: em percentual
df["Country"].value_counts(normalize=True)
```

```python
# Vamos criar intervalos de idade para entender melhor nossos usurios?
# Iremos criar categorias de idades com o intervalo de 4 em 4 anos.
# ou seja, 20-24, 25-29, 30-34, 35-39, 40-44, 45-49, 50-54, 55-59, +60

df.loc[df['Age'] < 25, 'grupo_idade'] = '20-24'
df.loc[(df['Age'] >= 25) & (df['Age'] < 30), 'grupo_idade'] = '25-29'
df.loc[(df['Age'] >= 30) & (df['Age'] < 35), 'grupo_idade'] = '30-34'
df.loc[(df['Age'] >= 35) & (df['Age'] < 40), 'grupo_idade'] = '35-39'
df.loc[(df['Age'] >= 40) & (df['Age'] < 45), 'grupo_idade'] = '40-44'
df.loc[(df['Age'] >= 45) & (df['Age'] < 50), 'grupo_idade'] = '45-49'
df.loc[(df['Age'] >= 50) & (df['Age'] < 55), 'grupo_idade'] = '50-54'
df.loc[(df['Age'] >= 55) & (df['Age'] < 60), 'grupo_idade'] = '55-59'
df.loc[df['Age'] >= 60, 'grupo_idade'] = '60+'
df
```

```python
# com o grupo de idades feito, qual ser a faixa etria mais recorrente?
df["grupo_idade"].value_counts()
```

# Visualizao dos dados


```python
# importando os pacotes que iremos utilizar
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# definir tamanho da figura
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como as faixas etrias
sns.countplot(data = df,
              x = 'grupo_idade')

# definir texto do ttulo e eixos x e y
plt.title('Contagem de faixas etrias')
plt.xlabel('Faixa etria')
plt.ylabel('Contagem')
plt.show()
```

```python
# definir tamanho da figura
plt.figure(figsize=(8, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos de plano
sns.countplot(data = df,
              x = 'Subscription Type')

# definir texto do ttulo e eixos x e y
plt.title('Contagem dos tipos de plano de assinatura')
plt.xlabel('Plano de assinatura')
plt.ylabel('Contagem')

plt.show()
```

```python
# definir tamanho da figura
plt.figure(figsize=(10, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os tipos
# de plano com tonalidades das barras definidas pela faixa etria

sns.countplot(data = df,
              x = 'Subscription Type',
              hue = 'grupo_idade')

# definir texto do ttulo e eixos x e y

plt.title('Planos de assinatura por faixa etria')
plt.xlabel('Faixa etria')
plt.ylabel('Contagem')

# plotar e definir texto da legenda
plt.legend(title = 'Faixa etria')

plt.show()
```

```python
# definir tamanho da figura
plt.figure(figsize=(12, 6))

# instanciar o tipo de plotagem, no caso, um countplot usando x como os pases
# com tonalidades das barras definidas pela faixa etria

sns.countplot(data = df,
              x = 'Country',
              hue = 'grupo_idade')

# definir texto do ttulo e eixos x e y

plt.title('Pases por faixa etria')
plt.xlabel('Pas')
plt.ylabel('Contagem')

# como o texto do eixo x  longo, podemos rotacion-lo para melhorar a visualizao
plt.xticks(rotation=330)

# plotar e definir texto da legenda
plt.legend(title = 'Faixa etria')
plt.show()
```
